```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
data_model = pd.read_csv("./output/data_model_def.csv")
data_model.head()
```

```python
data_model.max()
```

```python
scaler = StandardScaler()
```

```python
data_model["arrival_date_month"] = scaler.fit_transform(data_model["arrival_date_month"].values.reshape(-1,1))
data_model["arrival_date_week_number"] = scaler.fit_transform(data_model["arrival_date_week_number"].values.reshape(-1,1))
data_model["arrival_date_day_of_month"] = scaler.fit_transform(data_model["arrival_date_day_of_month"].values.reshape(-1,1))
data_model["stays_in_weekend_nights"] = scaler.fit_transform(data_model["stays_in_weekend_nights"].values.reshape(-1,1))
data_model["stays_in_week_nights"] = scaler.fit_transform(data_model["stays_in_week_nights"].values.reshape(-1,1))
data_model["adults"] = scaler.fit_transform(data_model["adults"].values.reshape(-1,1))
data_model["babies"] = scaler.fit_transform(data_model["babies"].values.reshape(-1,1))
data_model["children"] = scaler.fit_transform(data_model["children"].values.reshape(-1,1))
data_model["reserved_room_type"] = scaler.fit_transform(data_model["reserved_room_type"].values.reshape(-1,1))
data_model["total_of_special_requests"] = scaler.fit_transform(data_model["total_of_special_requests"].values.reshape(-1,1))
data_model["meal"] = scaler.fit_transform(data_model["meal"].values.reshape(-1,1))
```

```python
data_model.max()
```

```python
drop_cols = ["market_segment","distribution_channel","deposit_type","customer_type" ]
data_model = data_model.drop(drop_cols, axis =1)
data_model
```

```python
columnas_x = [a for a in (list(data_model.columns)) if a != "adr"]
columnas_x
```

```python
X = data_model[columnas_x]
X.head()
```

```python
y = data_model.adr
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 6500)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
```

```python
rdm = RandomForestRegressor()
```

```python
rdm.fit(X_train, y_train)
```

```python
y_pred = rdm.predict(X_test)
```

```python
rmse = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
rmse
```

```python
R2 = metrics.r2_score(y_test,y_pred)
R2
```
