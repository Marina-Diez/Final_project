```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

We import our data set an train our model by doing the following things:
- Min Max of all columns than have values higher than 1
- Drop not usefull columns
- We get a R2 = 0.8328084061904084

```python
data_model = pd.read_csv("./output/data_model_def.csv")
data_model.head()
```

```python
data_model.max()
```

```python
min_max = MinMaxScaler()
```

```python
data_model["arrival_date_month"] = min_max.fit_transform(data_model[["arrival_date_month"]])
data_model["arrival_date_week_number"] = min_max.fit_transform(data_model[["arrival_date_week_number"]])
data_model["arrival_date_day_of_month"] = min_max.fit_transform(data_model[["arrival_date_day_of_month"]])
data_model["stays_in_weekend_nights"] = min_max.fit_transform(data_model[["stays_in_weekend_nights"]])
data_model["stays_in_week_nights"] = min_max.fit_transform(data_model[["stays_in_week_nights"]])
data_model["adults"] = min_max.fit_transform(data_model[["adults"]])
data_model["babies"] = min_max.fit_transform(data_model[["babies"]])
data_model["children"] = min_max.fit_transform(data_model[["children"]])
data_model["reserved_room_type"] = min_max.fit_transform(data_model[["reserved_room_type"]])
data_model["total_of_special_requests"] = min_max.fit_transform(data_model[["total_of_special_requests"]])
data_model["meal"] = min_max.fit_transform(data_model[["meal"]])
```

```python
data_model.max()
```

```python
drop_cols = ["market_segment","distribution_channel","deposit_type","customer_type","arrival_date_week_number", "country" ]
data_model = data_model.drop(drop_cols, axis =1)
data_model
```

```python
columnas_x = [a for a in (list(data_model.columns)) if a != "adr"]
columnas_x
```

```python
X = data_model[columnas_x]
X.head()
```

```python
y = data_model.adr
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 6500)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
```

```python
rdm = RandomForestRegressor()
```

```python
rdm.fit(X_train, y_train)
```

```python
y_pred = rdm.predict(X_test)
```

```python
rmse = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
rmse
```

```python
R2 = metrics.r2_score(y_test,y_pred)
R2
```
